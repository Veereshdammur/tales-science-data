# Activation functions

An overview of the common activation functions used in working with neural networks. An activation is meant to be a non-linear function, because you want to destroy the linearity introduced by the transfer function alone, in order to make the network learn interesting patterns. In fact, the all purpose of the activation is to insert non-linearity in the model; without it you would just end up having stacks of linear operations the layers perform, which would essentially be like having a single layer and no hierarchy, preventing the recognition of patterns.

## ReLU

A rectifier linear unit, or ReLU in short, is simply the function

$$
f(x) = \max (0, x)
$$

and is a very common choice for an activation function. In practice what it does is nullifying a negative argument, see the plot. The ReLU is the easiest thing to conceive to go past linearity.
